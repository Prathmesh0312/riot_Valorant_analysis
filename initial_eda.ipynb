```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("valorant_games_project.csv")
df["date"] = pd.to_datetime(df["date"])

sns.set(style="whitegrid")

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="outcome")
plt.title("Match Outcome Distribution")
plt.xlabel("Outcome")
plt.ylabel("Number of Matches")
plt.show()
```

```python
#kills
plt.figure(figsize=(6,4))
sns.histplot(df["kills"], bins=20, kde=True)
plt.title("Distribution of Kills per Match")
plt.xlabel("Kills")
plt.ylabel("Frequency")
plt.show()
```

```python
#avg combat score
plt.figure(figsize=(6,4))
sns.histplot(df["acs"], bins=20, kde=True)
plt.title("Distribution of Average Combat Score (ACS)")
plt.xlabel("ACS")
plt.ylabel("Frequency")
plt.show()
```

```python
#agent selection
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    x="agent",
    order=df["agent"].value_counts().index
)
plt.title("Matches Played per Agent")
plt.xlabel("Agent")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()
```

```python
#frequency of map played
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    x="map",
    order=df["map"].value_counts().index
)
plt.title("Matches Played per Map")
plt.xlabel("Map")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()
    
```

```python
#KDA
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="outcome", y="kdr")
plt.title("KDR by Match Outcome")
plt.xlabel("Outcome")
plt.ylabel("KDR")
plt.show()
```

```python
#headshots
plt.figure(figsize=(6,4))
sns.histplot(df["headshot_pct"], bins=20, kde=True)
plt.title("Headshot Percentage Distribution")
plt.xlabel("Headshot %")
plt.ylabel("Frequency")
plt.show()
```
